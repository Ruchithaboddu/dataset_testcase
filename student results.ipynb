```python
merged_df=pd.merge(df2,df3,on="rno")
merged_df
```

```python
#problem statement:Analysis the given datasets and print the student names who have successfully cleared the exam.
pased_student=merged_df[merged_df["results"]=='pass']["names"]
print("student have successfully cleared  the exam:",list(pased_student))
```

```python
#problem 2:count no.of student(pass/fail) 
pass_student=merged_df[merged_df["results"]=='pass']
fail_student=merged_df[merged_df["results"]=='fail']
print("no.of student pass:",pass_student)
print("no.of student fail:",fail_student)
```
